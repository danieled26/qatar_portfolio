````python
"""
================================================================================
QATAR AIRWAYS — NLP SENTIMENT ANALYSIS PIPELINE
Ollama (local) | 32GB RAM | Run in VS Code
Senior Data Specialist Portfolio — Product Development & Design Team
================================================================================

SETUP (run these in a Colab cell before this script):
    pip install ollama pandas matplotlib seaborn wordcloud

GROQ API KEY:
    1. Sign up free at https://console.groq.com
    2. Create an API key
    3. Paste it into GROQ_API_KEY below (or use Colab Secrets)

GOOGLE DRIVE (recommended — saves progress across disconnects):
    from google.colab import drive
    drive.mount('/content/drive')
    OUTPUT_DIR = '/content/drive/MyDrive/qatar_nlp/'

OUTPUT FILES:
    qatar_sentiment_raw.json        — one record per review, all extracted fields
    qatar_sentiment_flat.csv        — flattened CSV for Power BI
    qatar_wordcloud_positive.png    — word cloud of positive sentiment terms
    qatar_wordcloud_negative.png    — word cloud of negative sentiment terms
    qatar_sentiment_trends.png      — sentiment over time charts
================================================================================
"""

# ── IMPORTS ───────────────────────────────────────────────────────────────────
import os
import json
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
from wordcloud import WordCloud
import ollama
from datetime import datetime
````

````python
# ── CONFIGURATION ─────────────────────────────────────────────────────────────
# No API key needed — Ollama runs locally
# Recommended models for 32GB RAM + 4GB GPU:
#   qwen2.5:7b   — best JSON adherence, fastest
#   phi4         — deepest reasoning, ~2x slower
#   llama3.1:8b  — solid all-rounder
MODEL         = "qwen2.5:7b"               # change to phi4 or llama3.1:8b if preferred
BATCH_SAVE    = 25                          # save progress every N reviews
OUTPUT_DIR    = "NLP_output"                        # change to Drive path in Colab
DATA_PATH     = "qatar_airline_reviews.csv" # adjust path as needed
SAMPLE_N      = None                        # set to int (e.g. 300) to sample, None = all

# Qatar Airways brand palette
C = {
    "maroon":     "#8D1B3D",
    "steel_dark": "#4A6278",
    "steel_mid":  "#6B8296",
    "grey_dark":  "#5A5A5A",
    "grey_light": "#C8C9CA",
    "off_white":  "#F7F5F2",
    "positive":   "#2E7D5A",
    "negative":   "#C0392B",
    "warning":    "#C47B00",
}

os.makedirs(OUTPUT_DIR, exist_ok=True)
# Ollama runs as a local service — no client init needed
````

````python
# ── EXTRACTION SCHEMA ─────────────────────────────────────────────────────────
# This is the heart of the pipeline.
# Phi-4 / Llama will return this exact JSON structure for every review.
SYSTEM_PROMPT = """
You are a senior aviation product analyst extracting structured insights from
Qatar Airways passenger reviews. Extract ONLY what is explicitly stated or
strongly implied in the review. Do not invent or infer beyond the text.

Return ONLY valid JSON. No explanation, no preamble, no markdown fences.
"""

def build_user_prompt(review: dict) -> str:
    return f"""
Analyse this Qatar Airways passenger review and return a JSON object with
exactly these fields:

{{
  "sentiment_overall": "positive" | "negative" | "neutral" | "mixed",
  "sentiment_score": <float -1.0 to 1.0, where -1=very negative, 1=very positive>,
  "emotion_primary": "satisfied" | "frustrated" | "delighted" | "disappointed" | "angry" | "neutral" | "impressed",
  "is_sarcastic": true | false,
  "sarcasm_flag_reason": "<brief explanation if sarcastic, else null>",

  "product_themes": {{
    "seat_comfort": "positive" | "negative" | "neutral" | "not_mentioned",
    "cabin_crew":   "positive" | "negative" | "neutral" | "not_mentioned",
    "food_beverage":"positive" | "negative" | "neutral" | "not_mentioned",
    "ife":          "positive" | "negative" | "neutral" | "not_mentioned",
    "wifi":         "positive" | "negative" | "neutral" | "not_mentioned",
    "lounge":       "positive" | "negative" | "neutral" | "not_mentioned",
    "boarding":     "positive" | "negative" | "neutral" | "not_mentioned",
    "cleanliness":  "positive" | "negative" | "neutral" | "not_mentioned",
    "value":        "positive" | "negative" | "neutral" | "not_mentioned",
    "punctuality":  "positive" | "negative" | "neutral" | "not_mentioned"
  }},

  "top_complaint": "<single most significant complaint in 8 words or less, or null>",
  "top_praise":    "<single most significant praise in 8 words or less, or null>",

  "sentiment_keywords": {{
    "positive_words": ["<up to 6 key positive words or short phrases from the review>"],
    "negative_words": ["<up to 6 key negative words or short phrases from the review>"]
  }},

  "competitor_mentioned": true | false,
  "competitor_name": "<competitor airline name if mentioned, else null>",
  "competitor_sentiment": "favourable_vs_qa" | "unfavourable_vs_qa" | "neutral" | null,

  "would_rebook_signal": "yes" | "no" | "maybe" | "not_stated",
  "service_recovery_mentioned": true | false,
  "loyalty_program_mentioned": true | false,

  "product_implications": [
    "<actionable product or service implication for QA product team, max 12 words>",
    "<second implication if applicable, else omit>"
  ]
}}

REVIEW CONTEXT:
- Date:         {review['review_date_iso']}
- Cabin:        {review['seat_type']}
- Traveller:    {review['type_of_traveller']}
- Route:        {review['route']}
- Aircraft:     {review['aircraft']}
- Star rating:  {review['overall_rating_10']} / 10
- Recommended:  {review['recommended']}
- Title:        {review['review_title']}

REVIEW TEXT:
{review['review_text']}
"""

# ── LOAD DATA ─────────────────────────────────────────────────────────────────
print("Loading reviews...")
df = pd.read_csv(DATA_PATH)
df["review_date_iso"] = pd.to_datetime(df["review_date_iso"], errors="coerce")
df["year"]  = df["review_date_iso"].dt.year
df["month"] = df["review_date_iso"].dt.month

# Fill nulls for context fields
for col in ["seat_type","type_of_traveller","route","aircraft",
            "overall_rating_10","recommended","review_title","review_text"]:
    df[col] = df[col].fillna("Not stated")

# Optional stratified sample
if SAMPLE_N:
    # Stratify by cabin class and rating band to ensure coverage
    df["rating_band"] = pd.cut(
        pd.to_numeric(df["overall_rating_10"], errors="coerce"),
        bins=[0,3,6,8,10],
        labels=["low","mid","high","top"]
    )
    df = (df.groupby(["seat_type","rating_band"], observed=True)
            .apply(lambda x: x.sample(
                min(len(x), max(1, SAMPLE_N // 12)),
                random_state=42))
            .reset_index(drop=True))
    print(f"  Stratified sample: {len(df)} reviews")
else:
    print(f"  Full dataset: {len(df)} reviews")


# ── LOAD EXISTING PROGRESS ────────────────────────────────────────────────────
RESULTS_PATH = os.path.join(OUTPUT_DIR, "qatar_sentiment_raw.json")
results = []
processed_ids = set()

if os.path.exists(RESULTS_PATH):
    with open(RESULTS_PATH, "r") as f:
        results = json.load(f)
    processed_ids = {r["review_id"] for r in results}
    print(f"  Resuming: {len(processed_ids)} reviews already processed")

# ── EXTRACTION LOOP ───────────────────────────────────────────────────────────
print(f"\nStarting extraction with {MODEL}...")
print(f"Saving progress every {BATCH_SAVE} reviews to {RESULTS_PATH}\n")

errors = 0

for idx, row in df.iterrows():

    review_id = str(idx)
    if review_id in processed_ids:
        continue

    review = row.to_dict()

    try:
        response = ollama.chat(
            model=MODEL,
            messages=[
                {"role": "system", "content": SYSTEM_PROMPT},
                {"role": "user",   "content": build_user_prompt(review)},
            ],
            options={"temperature": 0.1},   # low temp = consistent structured output
        )

        raw_text = response["message"]["content"].strip()

        # Strip markdown fences if model adds them despite instructions
        if raw_text.startswith("```"):
            raw_text = raw_text.split("```")[1]
            if raw_text.startswith("json"):
                raw_text = raw_text[4:]

        extracted = json.loads(raw_text)

        # Merge extracted fields with original review metadata
        record = {
            # ── Original review fields ─────────────────────────────────────
            "review_id":            review_id,
            "review_date":          str(row["review_date_iso"])[:10],
            "year":                 int(row["year"]) if pd.notna(row["year"]) else None,
            "month":                int(row["month"]) if pd.notna(row["month"]) else None,
            "reviewer_country":     row["reviewer_country"],
            "seat_type":            row["seat_type"],
            "type_of_traveller":    row["type_of_traveller"],
            "route":                row["route"],
            "aircraft":             row["aircraft"],
            "overall_rating_10":    row["overall_rating_10"],
            "recommended":          row["recommended"],
            "trip_verified":        row["trip_verified"],
            "review_title":         row["review_title"],
            "review_text":          row["review_text"],
            # Original numeric scores
            "score_seat_comfort":           row.get("seat_comfort"),
            "score_cabin_staff":            row.get("cabin_staff_service"),
            "score_food_beverages":         row.get("food_beverages"),
            "score_ife":                    row.get("inflight_entertainment"),
            "score_ground_service":         row.get("ground_service"),
            "score_wifi":                   row.get("wifi_connectivity"),
            "score_value_for_money":        row.get("value_for_money"),
            # ── LLM extracted fields ───────────────────────────────────────
            **extracted,
        }

        results.append(record)
        processed_ids.add(review_id)

        # Progress log
        n = len(results)
        print(f"  [{n}/{len(df)}] {row['seat_type'][:3]} | "
              f"{str(row['review_date_iso'])[:10]} | "
              f"{extracted.get('sentiment_overall','?'):8s} | "
              f"score={extracted.get('sentiment_score','?')}")

        # Incremental save
        if n % BATCH_SAVE == 0:
            with open(RESULTS_PATH, "w") as f:
                json.dump(results, f, indent=2, default=str)
            print(f"  ── Progress saved ({n} reviews) ──")

        # No rate limit locally — small sleep keeps CPU from pegging
        time.sleep(0.1)

    except json.JSONDecodeError as e:
        errors += 1
        print(f"  ✗ JSON parse error on review {idx}: {e}")
        print(f"    Raw response: {raw_text[:200]}")
        time.sleep(3)

    except Exception as e:
        errors += 1
        print(f"  ✗ Error on review {idx}: {e}")
        time.sleep(5)

# Final save
with open(RESULTS_PATH, "w") as f:
    json.dump(results, f, indent=2, default=str)
print(f"\nExtraction complete: {len(results)} reviews | {errors} errors")



# ── LOAD EXISTING PROGRESS ────────────────────────────────────────────────────
RESULTS_PATH = os.path.join(OUTPUT_DIR, "qatar_sentiment_raw.json")
results = []
processed_ids = set()

if os.path.exists(RESULTS_PATH):
    with open(RESULTS_PATH, "r") as f:
        results = json.load(f)
    processed_ids = {r["review_id"] for r in results}
    print(f"  Resuming: {len(processed_ids)} reviews already processed")

# ── EXTRACTION LOOP ───────────────────────────────────────────────────────────
print(f"\nStarting extraction with {MODEL}...")
print(f"Saving progress every {BATCH_SAVE} reviews to {RESULTS_PATH}\n")

errors = 0

for idx, row in df.iterrows():

    review_id = str(idx)
    if review_id in processed_ids:
        continue

    review = row.to_dict()

    try:
        response = ollama.chat(
            model=MODEL,
            messages=[
                {"role": "system", "content": SYSTEM_PROMPT},
                {"role": "user",   "content": build_user_prompt(review)},
            ],
            options={"temperature": 0.1},   # low temp = consistent structured output
        )

        raw_text = response["message"]["content"].strip()

        # Strip markdown fences if model adds them despite instructions
        if raw_text.startswith("```"):
            raw_text = raw_text.split("```")[1]
            if raw_text.startswith("json"):
                raw_text = raw_text[4:]

        extracted = json.loads(raw_text)

        # Merge extracted fields with original review metadata
        record = {
            # ── Original review fields ─────────────────────────────────────
            "review_id":            review_id,
            "review_date":          str(row["review_date_iso"])[:10],
            "year":                 int(row["year"]) if pd.notna(row["year"]) else None,
            "month":                int(row["month"]) if pd.notna(row["month"]) else None,
            "reviewer_country":     row["reviewer_country"],
            "seat_type":            row["seat_type"],
            "type_of_traveller":    row["type_of_traveller"],
            "route":                row["route"],
            "aircraft":             row["aircraft"],
            "overall_rating_10":    row["overall_rating_10"],
            "recommended":          row["recommended"],
            "trip_verified":        row["trip_verified"],
            "review_title":         row["review_title"],
            "review_text":          row["review_text"],
            # Original numeric scores
            "score_seat_comfort":           row.get("seat_comfort"),
            "score_cabin_staff":            row.get("cabin_staff_service"),
            "score_food_beverages":         row.get("food_beverages"),
            "score_ife":                    row.get("inflight_entertainment"),
            "score_ground_service":         row.get("ground_service"),
            "score_wifi":                   row.get("wifi_connectivity"),
            "score_value_for_money":        row.get("value_for_money"),
            # ── LLM extracted fields ───────────────────────────────────────
            **extracted,
        }

        results.append(record)
        processed_ids.add(review_id)

        # Progress log
        n = len(results)
        print(f"  [{n}/{len(df)}] {row['seat_type'][:3]} | "
              f"{str(row['review_date_iso'])[:10]} | "
              f"{extracted.get('sentiment_overall','?'):8s} | "
              f"score={extracted.get('sentiment_score','?')}")

        # Incremental save
        if n % BATCH_SAVE == 0:
            with open(RESULTS_PATH, "w") as f:
                json.dump(results, f, indent=2, default=str)
            print(f"  ── Progress saved ({n} reviews) ──")

        # No rate limit locally — small sleep keeps CPU from pegging
        time.sleep(0.1)

    except json.JSONDecodeError as e:
        errors += 1
        print(f"  ✗ JSON parse error on review {idx}: {e}")
        print(f"    Raw response: {raw_text[:200]}")
        time.sleep(3)

    except Exception as e:
        errors += 1
        print(f"  ✗ Error on review {idx}: {e}")
        time.sleep(5)

# Final save
with open(RESULTS_PATH, "w") as f:
    json.dump(results, f, indent=2, default=str)
print(f"\nExtraction complete: {len(results)} reviews | {errors} errors")
````

````python

# ── FLATTEN TO CSV FOR POWER BI ───────────────────────────────────────────────
print("\nFlattening to CSV for Power BI...")

flat_rows = []
for r in results:
    flat = {
        # Identity
        "review_id":              r.get("review_id"),
        "review_date":            r.get("review_date"),
        "year":                   r.get("year"),
        "month":                  r.get("month"),
        "reviewer_country":       r.get("reviewer_country"),
        "seat_type":              r.get("seat_type"),
        "type_of_traveller":      r.get("type_of_traveller"),
        "route":                  r.get("route"),
        "aircraft":               r.get("aircraft"),
        "overall_rating_10":      r.get("overall_rating_10"),
        "recommended":            r.get("recommended"),
        "trip_verified":          r.get("trip_verified"),
        "review_title":           r.get("review_title"),

        # Original numeric scores
        "score_seat_comfort":     r.get("score_seat_comfort"),
        "score_cabin_staff":      r.get("score_cabin_staff"),
        "score_food_beverages":   r.get("score_food_beverages"),
        "score_ife":              r.get("score_ife"),
        "score_ground_service":   r.get("score_ground_service"),
        "score_wifi":             r.get("score_wifi"),
        "score_value_for_money":  r.get("score_value_for_money"),

        # LLM sentiment fields
        "sentiment_overall":      r.get("sentiment_overall"),
        "sentiment_score":        r.get("sentiment_score"),
        "emotion_primary":        r.get("emotion_primary"),
        "is_sarcastic":           r.get("is_sarcastic"),
        "sarcasm_flag_reason":    r.get("sarcasm_flag_reason"),

        # Product themes (flattened)
        "theme_seat_comfort":     r.get("product_themes", {}).get("seat_comfort"),
        "theme_cabin_crew":       r.get("product_themes", {}).get("cabin_crew"),
        "theme_food_beverage":    r.get("product_themes", {}).get("food_beverage"),
        "theme_ife":              r.get("product_themes", {}).get("ife"),
        "theme_wifi":             r.get("product_themes", {}).get("wifi"),
        "theme_lounge":           r.get("product_themes", {}).get("lounge"),
        "theme_boarding":         r.get("product_themes", {}).get("boarding"),
        "theme_cleanliness":      r.get("product_themes", {}).get("cleanliness"),
        "theme_value":            r.get("product_themes", {}).get("value"),
        "theme_punctuality":      r.get("product_themes", {}).get("punctuality"),

        # Key signals
        "top_complaint":          r.get("top_complaint"),
        "top_praise":             r.get("top_praise"),
        "competitor_mentioned":   r.get("competitor_mentioned"),
        "competitor_name":        r.get("competitor_name"),
        "competitor_sentiment":   r.get("competitor_sentiment"),
        "would_rebook_signal":    r.get("would_rebook_signal"),
        "service_recovery_mentioned": r.get("service_recovery_mentioned"),
        "loyalty_program_mentioned":  r.get("loyalty_program_mentioned"),

        # Keywords joined as pipe-separated strings for Power BI splitting
        "positive_keywords":      " | ".join(r.get("sentiment_keywords", {}).get("positive_words", [])),
        "negative_keywords":      " | ".join(r.get("sentiment_keywords", {}).get("negative_words", [])),

        # Product implications joined
        "product_implications":   " | ".join(map(str, r.get("product_implications") or [])), #join(str(item) if item is not None else "" for item in data)
    }
    flat_rows.append(flat)

flat_df = pd.DataFrame(flat_rows)
CSV_PATH = os.path.join(OUTPUT_DIR, "qatar_sentiment_flat.csv")
flat_df.to_csv(CSV_PATH, index=False)
print(f"  ✓ Saved {CSV_PATH} ({len(flat_df)} rows × {len(flat_df.columns)} columns)")


# ── WORD CLOUDS ───────────────────────────────────────────────────────────────
print("\nGenerating word clouds...")

def make_wordcloud(words_list, title, filename, colormap, bg_colour):
    text = " ".join([w for sublist in words_list for w in sublist if w])
    if not text.strip():
        print(f"  ✗ No words for {title}")
        return
    wc = WordCloud(
        width=1200, height=600,
        background_color=bg_colour,
        colormap=colormap,
        max_words=80,
        collocations=False,
        prefer_horizontal=0.85,
        min_font_size=10,
    ).generate(text)

    fig, ax = plt.subplots(figsize=(14, 7))
    fig.patch.set_facecolor(C["off_white"])
    ax.imshow(wc, interpolation="bilinear")
    ax.axis("off")
    ax.set_title(title, fontsize=16, fontweight="bold",
                 color=C["maroon"] if "Positive" in title else C["negative"],
                 pad=16)
    fig.tight_layout()
    path = os.path.join(OUTPUT_DIR, filename)
    fig.savefig(path, dpi=180, bbox_inches="tight", facecolor=C["off_white"])
    plt.close(fig)
    print(f"  ✓ Saved {path}")

# Build word lists from results
pos_words = [r.get("sentiment_keywords", {}).get("positive_words", []) for r in results]
neg_words = [r.get("sentiment_keywords", {}).get("negative_words", []) for r in results]

# Maroon-toned colormap for negative, teal for positive
make_wordcloud(pos_words, "Positive Sentiment — Key Words & Phrases",
               "qatar_wordcloud_positive.png", "YlGn",  C["off_white"])
make_wordcloud(neg_words, "Negative Sentiment — Key Words & Phrases",
               "qatar_wordcloud_negative.png", "OrRd",  C["off_white"])


# ── SENTIMENT TREND CHARTS ────────────────────────────────────────────────────
print("\nGenerating sentiment trend charts...")

adf = flat_df.copy()
adf["overall_rating_10"] = pd.to_numeric(adf["overall_rating_10"], errors="coerce")
adf["sentiment_score"]   = pd.to_numeric(adf["sentiment_score"],   errors="coerce")
adf["year"]              = pd.to_numeric(adf["year"], errors="coerce")
adf = adf[adf["year"].notna() & (adf["year"] >= 2016)]

plt.rcParams.update({
    "figure.facecolor":  C["off_white"],
    "axes.facecolor":    C["off_white"],
    "axes.spines.top":   False,
    "axes.spines.right": False,
    "axes.grid":         True,
    "grid.alpha":        0.2,
    "grid.linestyle":    "--",
    "grid.color":        C["grey_light"],
    "font.size":         11,
})

fig = plt.figure(figsize=(20, 14))
fig.suptitle(
    "Qatar Airways — NLP Sentiment Analysis | Groq + Llama 3.1\n"
    "Skytrax Reviews 2016–2026",
    fontsize=16, fontweight="bold", y=1.01,
    color=C["maroon"]
)
gs = gridspec.GridSpec(3, 3, figure=fig, hspace=0.5, wspace=0.4)

years = sorted(adf["year"].dropna().unique().astype(int))

# Chart 1 — Sentiment score trend vs star rating (dual axis)
ax = fig.add_subplot(gs[0, :2])
yr_sent   = adf.groupby("year")["sentiment_score"].mean()
yr_rating = adf.groupby("year")["overall_rating_10"].mean()
ax2 = ax.twinx()
ax.plot(yr_sent.index,   yr_sent.values,   "o-",
        color=C["maroon"],     linewidth=2.5, markersize=7, label="LLM Sentiment Score (left)")
ax2.plot(yr_rating.index, yr_rating.values, "s--",
        color=C["steel_dark"], linewidth=2,   markersize=6, label="Star Rating avg (right)", alpha=0.8)
ax.set_ylabel("Avg Sentiment Score (-1 to 1)", color=C["maroon"])
ax2.set_ylabel("Avg Star Rating (1-10)",        color=C["steel_dark"])
ax.set_title("LLM Sentiment Score vs Star Rating — Yearly Trend", fontweight="bold")
ax.set_xlabel("Year"); ax.set_xticks(years)
ax.axhline(0, color=C["grey_dark"], linestyle=":", linewidth=1)
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, fontsize=9)

# Chart 2 — Sentiment distribution donut
ax = fig.add_subplot(gs[0, 2])
sent_counts = adf["sentiment_overall"].value_counts()
sent_colours = {
    "positive": C["positive"], "negative": C["negative"],
    "neutral":  C["steel_mid"],"mixed":    C["warning"],
}
colours_pie = [sent_colours.get(k, C["grey_dark"]) for k in sent_counts.index]
ax.pie(sent_counts, labels=sent_counts.index, colors=colours_pie,
       autopct="%1.1f%%", startangle=90,
       wedgeprops={"width": 0.55, "edgecolor": "white", "linewidth": 2},
       pctdistance=0.75)
ax.set_title("Overall Sentiment Distribution", fontweight="bold")

# Chart 3 — Sentiment by cabin class
ax = fig.add_subplot(gs[1, 0])
cabin_sent = adf.groupby("seat_type")["sentiment_score"].mean().sort_values()
cabin_colours = ["#C0392B" if v < 0 else "#2E7D5A" if v > 0.3 else "#C47B00"
                 for v in cabin_sent.values]
bars = ax.barh(cabin_sent.index, cabin_sent.values,
               color=cabin_colours, edgecolor="none", height=0.55)
ax.axvline(0, color=C["grey_dark"], linewidth=1)
ax.set_title("Avg Sentiment Score by Cabin", fontweight="bold")
ax.set_xlabel("Sentiment Score (-1 to 1)")

# Chart 4 — Sentiment by traveller type
ax = fig.add_subplot(gs[1, 1])
trav_sent = adf.groupby("type_of_traveller")["sentiment_score"].mean().sort_values()
trav_cols  = ["#C0392B" if v < 0 else "#2E7D5A" if v > 0.3 else "#C47B00"
              for v in trav_sent.values]
ax.barh(trav_sent.index, trav_sent.values,
        color=trav_cols, edgecolor="none", height=0.55)
ax.axvline(0, color=C["grey_dark"], linewidth=1)
ax.set_title("Avg Sentiment Score by Traveller Type", fontweight="bold")
ax.set_xlabel("Sentiment Score (-1 to 1)")

# Chart 5 — Emotion distribution
ax = fig.add_subplot(gs[1, 2])
emotion_counts = adf["emotion_primary"].value_counts()
emotion_cols = {
    "delighted":     C["positive"],
    "satisfied":     "#5A9E7A",
    "impressed":     C["steel_mid"],
    "neutral":       C["grey_light"],
    "disappointed":  C["warning"],
    "frustrated":    "#D4730A",
    "angry":         C["negative"],
}
e_colours = [emotion_cols.get(e, C["grey_dark"]) for e in emotion_counts.index]
ax.bar(emotion_counts.index, emotion_counts.values,
       color=e_colours, edgecolor="none", width=0.65)
ax.set_title("Primary Emotion Distribution", fontweight="bold")
ax.set_xlabel("Emotion"); ax.set_ylabel("Count")
ax.set_xticklabels(emotion_counts.index, rotation=30, ha="right")

# Chart 6 — Product theme negativity heatmap (which products draw negative sentiment)
ax = fig.add_subplot(gs[2, :2])
theme_cols = [c for c in adf.columns if c.startswith("theme_")]
theme_labels = [c.replace("theme_", "").replace("_", " ").title() for c in theme_cols]
cabins_plot = ["Economy Class", "Business Class", "First Class"]
heat_data = []
for cabin in cabins_plot:
    sub = adf[adf["seat_type"] == cabin]
    row_vals = []
    for col in theme_cols:
        neg_pct = (sub[col] == "negative").sum() / max(len(sub[sub[col] != "not_mentioned"]), 1) * 100
        row_vals.append(neg_pct)
    heat_data.append(row_vals)

heat_df = pd.DataFrame(heat_data, index=cabins_plot, columns=theme_labels)
QA_RED_CMAP = LinearSegmentedColormap.from_list("qa_red", [C["off_white"], C["warning"], C["negative"]])
sns.heatmap(heat_df, ax=ax, cmap=QA_RED_CMAP, vmin=0, vmax=60,
            annot=True, fmt=".0f", annot_kws={"size": 9},
            linewidths=0.4, cbar_kws={"label": "% Negative Mentions", "shrink": 0.8})
ax.set_title("Product Theme Negativity Rate (%) — Cabin × Theme", fontweight="bold")
ax.set_xticklabels(ax.get_xticklabels(), rotation=35, ha="right", fontsize=9)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

# Chart 7 — Sarcasm detection
ax = fig.add_subplot(gs[2, 2])
sarc_counts = adf["is_sarcastic"].value_counts()
ax.pie(sarc_counts,
       labels=["Not sarcastic", "Sarcastic"],
       colors=[C["steel_mid"], C["maroon"]],
       autopct="%1.1f%%", startangle=90,
       wedgeprops={"edgecolor": "white", "linewidth": 2})
total_sarc = sarc_counts.get(True, 0)
ax.set_title(f"Sarcasm Detection\n({total_sarc} reviews flagged)", fontweight="bold")

fig.text(0.99, 0.01,
         "Qatar Airways | NLP Pipeline: Groq + Llama 3.1 8B | Skytrax Reviews 2016–2026",
         ha="right", va="bottom", fontsize=8, color=C["grey_dark"], alpha=0.6)

TREND_PATH = os.path.join(OUTPUT_DIR, "qatar_sentiment_trends.png")
fig.savefig(TREND_PATH, dpi=180, bbox_inches="tight", facecolor=C["off_white"])
plt.close(fig)
print(f"  ✓ Saved {TREND_PATH}")


# ── SUMMARY PRINT ─────────────────────────────────────────────────────────────
print("\n" + "=" * 60)
print("PIPELINE COMPLETE")
print("=" * 60)
print(f"  Reviews processed : {len(results)}")
print(f"  Errors            : {errors}")
print()
print("Output files:")
print(f"  qatar_sentiment_raw.json         ← full JSON, one record per review")
print(f"  qatar_sentiment_flat.csv         ← Power BI ready, {len(flat_df.columns)} columns")
print(f"  qatar_wordcloud_positive.png     ← positive sentiment word cloud")
print(f"  qatar_wordcloud_negative.png     ← negative sentiment word cloud")
print(f"  qatar_sentiment_trends.png       ← 7-chart sentiment dashboard")
print()
print("Power BI — recommended visuals from qatar_sentiment_flat.csv:")
print("  • sentiment_score over review_date  (line chart)")
print("  • sentiment_overall by seat_type    (stacked bar)")
print("  • theme_* columns by year           (heatmap matrix)")
print("  • top_complaint word frequency      (bar chart)")
print("  • emotion_primary distribution      (donut)")
print("  • competitor_name mentions          (bar)")
print("  • would_rebook_signal by cabin      (stacked bar)")
````
